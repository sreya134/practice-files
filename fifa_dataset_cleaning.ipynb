###working on height and weight columns

```python
df['Height']=df['Height'].str.replace('cm','')
df.head()
```

```python
df['Weight']=df['Weight'].str.replace('kg','')
df.head()
```

```python
def height_to_integers(height):
    if isinstance(height, str):
        if "'" in height:
            feet, inches = height.split("'")
            feet_cm = int(feet) * 30.48
            inches_cm = int(inches.replace('"', '')) * 2.54
            total_cm = feet_cm + inches_cm
            return int(total_cm)
    return height
```

```python
df['Height'] = df['Height'].apply(height_to_integers)
```

```python
def weight_to_lala(weight):
  if isinstance(weight, str):
    if weight.find("lbs") != -1:
      weight = weight.replace("lbs", " ")
      weight = int(weight) * 0.4535
      return int(weight)
  else:
    return int(weight)
```

```python
df["Weight"] = df["Weight"].apply(weight_to_lala)
```

```python
df.head()
```

```python

df["SM"] = df["SM"].str.rstrip("★")
```

```python
#Value', 'Wage' and "Release Clause' are string columns. Convert them to numbers. For eg, "M" in value column is Million, so multiply the row values by 1,000,000, etc.

def ConvertMoney(money:str):
    if '€' in money:
        money=money.replace("€","")
    if 'M' in money:
        return int(float(money.replace('M',''))*1000000)
    elif 'K' in money:
        return int(float(money.replace('K',''))*1000)

df['Value'] = df['Value'].apply(ConvertMoney)
df['Wage'] = df['Wage'].apply(ConvertMoney)
df['Release Clause'] = df['Release Clause'].apply(ConvertMoney)
df[['Value','Wage','Release Clause']]
```
